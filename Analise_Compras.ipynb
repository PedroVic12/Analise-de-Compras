**SITUAÇÃO HIPOTÉTICA**

Uma empresa faz uma licitação para vender seus produtos, porem, existem vários fornecedores que oferecem esse mesmo produto e isso gera uma concorrência, pois esse fornecedor não é o único que oferece esse ou esses mesmos tipos de produtos. 

Cada fornecedor faz sua proposta e a **pessoa responsável pela compra** do produto tem que ficar comparando o preço **manualmente** de um produto de cada empresa e saber qual vai ter o melhor custo beneficio. 

Portando, cada fornecedor preencheu sua planilha do Excel e mostrando seus **itens, quantidade, valor unitário e o valor total do produto.**

O resultado foi 8 planilhas do excel com todas essas informações

A ideia desse Jupyter notebook, é mostrar que é possível automatizar esse processo através do Python e fazer uma analise.

    A ideia é usar o Python para:

    - Ler todos os arquivos de Excel

    - Fazer uma tabela única

    - Fazer uma analise Inicial

    - Pegar o mínimo de cada item em cada empresa(buscar um preço ideal)

    - Através desse preço, é possível negociar com o fornecedor.

    - No final, comparar o preço inicial e o preço final e perceber o quanto de economia foi utilizada nesse processo.

```python
import pandas as pd
import os
```

```python
arquivos = os.listdir()


fornecedores = arquivos[2:]
print(fornecedores)

print("\nMostrando os arquivos de Excel mandado pelos fornecedores")
```

### Lendo 1 planilha do excel

```python
licitacao = pd.read_excel("Empresa A.xlsx")
display(licitacao)
```

### Trocando o nome "Valor" dessa tabela para o nome da empresa, futuramente, isso vai nos ajudar.

### → Usando a empresa A como exemplo.

```python
licitacao = licitacao.rename(columns={'Total': 'Empresa A'})
display(licitacao)
```

### Retirando o Valor Unitário, pois não vai ser util no momento. Me interessa o valor Total de cada empresa apenas

```python
licitacao = licitacao.drop(['Valor Unitário'], axis =1)
display(licitacao)
```

### Lendo os demais arquivos de Excel e colocar na mesma tabela.

```python
for empresas in fornecedores:
    propostas_empresas = pd.read_excel(empresas)
    licitacao[empresas[:-5]] = propostas_empresas['Total'].copy()
display(licitacao)
```

### Verificando o Somatório de cada Valor Total de cada empresa

```python
print("Soma do valor total da empresa A: ")
print(licitacao['Empresa A'].sum(), "reais")
```

```python
for coluna in licitacao.columns[3:]:
    valor_total_cada_empresa = licitacao[coluna].sum()
    # print(valor_total_cada_empresa)
    print('Valor Total: ')
    print(f" {coluna} = {valor_total_cada_empresa:,.2f} reais  \n")
```


### So por essa **analise**, a **empresa B** ta com o valor **MUITO** abaixo em relação as outras empresas.

 - O fornecedor pode ter tido um erro manual ao fazer a tabela

 - Esse fornecedor pode não ser tao bom, podem ter equipamentos/produtos de qualidade não tao boa ou duvidosa.

 - Ou então, na melhor das hipóteses, esse fornecedor pode possuir os melhores preços.

### Buscando os menores valores de cada ITEM de cada Empresa

```python
licitacao['Valor Minimo'] = licitacao.drop(['Item', 'Descrição do Item', 'Quantidade'], axis = 1).min(axis=1)

soma_minimo = licitacao['Valor Minimo'].sum()

display(licitacao)


print(f"Somando todos os valores MINIMOS, a gente consegue o total de R${soma_minimo:,.2f} reais" )
```

### Esse seria o Preço ideal, então a ideia é sempre buscar um preço proximo dele.A ideia é usar esse preço como referencia para ser possível uma negociação

### Criando uma planilha com os desvios de cada fornecedor.

```python
for empresa in licitacao.columns[3:-1]:
    desvio = licitacao[['Item','Descrição do Item', 'Quantidade', empresa,'Valor Minimo']]
    desvio['Delta'] = desvio[empresa] - desvio['Valor Minimo']
    desvio.to_excel('Diferença - '+ empresa +'.xlsx',index=False)
```

<font color = blue>

# **Conclusão**

- Com as planilhas com os desvios criadas. É possível analisar uma margem de dinheiro de cada produto para ser negociado com o fornecedor de acordo com a preferência do Item.

- Foi encontrado um Valor Ideal de R$193.299,00 reais, então quanto mais proximo desse preço, melhor.

- Apos a compra dos Itens, o comprador pode analisar e perceber o quanto ele economizou através desse processo todo.

## Pegando a Empresa B, que era uma forte candidata.

```python
exemplo_df = pd.read_excel('Diferença - Empresa B.xlsx')
display(exemplo_df)
```

<font color = green>

### ``Valor mínimo``: É o valor mais barato na qual uma outra empresa concorrente cobre esse preço NO MESMO PRODUTO.

### ``Delta``: É o valor que falta do valor mínimo para o valor que a empresa B esta cobrando. 

### Ou seja, Seria possível negociar o valor através do desse numero de Delta.Sendo assim, voce possui um caminho mais simples e mais rápido para negociar e conseguir o produto desejado.
